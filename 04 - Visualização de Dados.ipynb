```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def configurar_matplotlib():
    plt.rcParams['figure.figsize'] = (20, 7)
```

```python
titanic = pd.read_csv('dados/train.csv')
titanic
```

```python
configurar_matplotlib()
```

```python
titanic.shape
```

```python
# Plot de teste para as principais funções da Matplotlib

plt.plot(titanic["Age"], "*-r")
plt.title("Informações dos passageiros pela idade", size = 14)
plt.xlabel("Passageiro", size = 14)
plt.ylabel("Idade", size = 14)
plt.show()
```

```python
# Plotando com a Matplotlib internamente na biblioteca Pandas, diretamente a partir do DataFrame

titanic["Age"].plot()
plt.title("Informações dos passageiros pela idade", size = 14)
plt.xlabel("Passageiro", size = 14)
plt.ylabel("Idade", size = 14)
plt.show()
```

```python
titanic.plot()
plt.show()
```

#### Plotando um gráfico de dispersão (Scatter Plot)

```python
plt.scatter(titanic["PassengerId"], titanic["Age"], color = "g")
plt.title("Dispersão dos passageiros pela idade", size = 14)
plt.xlabel("Passageiro", size = 14)
plt.ylabel("Idade", size = 14)
plt.show()
```

```python
titanic["Age"].describe()
```

```python
# Plotando um gráfico de histograma

titanic["Age"].hist()
plt.title("Histograma de Idade", size = 14)
plt.xlabel("Idade")
plt.ylabel("Frequência", size = 14)
plt.savefig("histograma.png", trasparent = True)
plt.show()
```
